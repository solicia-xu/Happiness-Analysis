#### This notebook takes a country code mapping database (countrycodes.csv) that are missing from our other mapping database (worldbank.csv) and applies it to our other datasource (world-happiness-report-2019.csv) so we can join on the iso2code (unique value) of a country rather than the country name

```python
import os
import pandas as pd
import sys
sys.path.insert(1, '../../notebooks/libs')
from api_call import worldbank
```

```python
try:
    worldbank_df = worldbank()
except:
    print("Oops!",sys.exc_info()[0],"occured, using CSV Instead")
    worldbank_df = pd.read_csv(os.path.join('worldbank.csv'))
    
worldhappiness_df = pd.read_csv(os.path.join('world-happiness-report-2019.csv'))
missingcodes = pd.read_csv(os.path.join('countrycodes.csv'))
cclookup = pd.concat([worldbank_df,missingcodes], sort=True)
```

```python
cclist = []

for i, row in worldhappiness_df.iterrows():
    cclist.append(cclookup['Country Code'][cclookup['Country']== row['Country name']].values[0])
#     print(row['Country name'])
#     print(cclookup['Country Code'][cclookup['Country']== row['Country name']])
# worldhappiness_df
```

```python
worldhappiness_df['Country Code'] = cclist
```

```python
worldhappiness_df.to_csv(os.path.join("worldhappiness_prepped.csv"), index=None)
```
