## Now we loop through our countries, making a url using country ID and retreiving our first data value, revenue in 2019

```python
CountryRevenue2019 = CountryRevenueInfo.copy(deep=True)
#prep columns
CountryRevenue2019['Year']="2019"
CountryRevenue2019['Revenue ($M)']=""
for index, row in CountryRevenueInfo.iterrows():
    #we create our url and call for the html, saving the text
    page = requests.get('https://www.statista.com/outlook/251/' + str(row['Country ID']) + '/consumer-electronics/worldwide')
    tree = html.fromstring(page.content)
    #here we search for the name of our div class to find the value we need, and save the text
    revenue = tree.xpath('//div[@class="font-size-xl text-light margin-top-5"]/text()')
    #some text editing to get it into the format we want
    word=revenue[0].split('$')
    money = word[1].split('m')
    #and we save our value
    CountryRevenue2019.loc[[index],['Revenue ($M)']] = money[0]
print (CountryRevenue2019)
```

## It is at this point where we run into another major issue: The values for 2017 and 2018 are stored in a table that we can't access by retrieving text from the html code. So we use BeautifulSoup and retrieve the code itself. We do some light text editing to retrieve the data from the code for the table. Conveniently, despite the fact that we need to have an account to actually view the data on the website, our html call bypasses this.

```python
#prepare all our columns
CountryRevenue2018 = CountryRevenueInfo.copy(deep=True)
CountryRevenue2017 = CountryRevenueInfo.copy(deep=True)
CountryRevenue2017['Year'] = '2017'
CountryRevenue2018['Year'] = '2018'
CountryRevenueInfo['Revenue ($M)']=""
CountryRevenueInfo['Revenue ($M)']=""
for index, row in CountryRevenueInfo.iterrows():
    #create our url to retrieve html with BeautifulSoup
    page = requests.get('https://www.statista.com/outlook/251/' + str(row['Country ID']) + '/consumer-electronics/worldwide')
    data = page.text
    soup = BeautifulSoup(data,features="lxml")
    #find the chart we want and retrieve its data
    theclass = soup.find("div",'cChart cContainer active revenue')
    theclass['data-highcharts']
    #do some text editing to find and save the values we want
    split1 = theclass['data-highcharts'].split('\"data\":[')
    split2 = split1[1].split(']}]')
    split3 = split2[0].split(',')
    CountryRevenue2017.loc[[index],['Revenue ($M)']] = split3[0]
    CountryRevenue2018.loc[[index],['Revenue ($M)']] = split3[1]
#print (CountryRevenueInfo)
```

```python
#combine our dataframes for each year into one dataframe and print it to a csv
CountryRevenueInfo = pd.concat([CountryRevenue2017,CountryRevenue2018,CountryRevenue2019], ignore_index = False, sort=True)

CountryRevenueInfo['Revenue ($M)'] = (CountryRevenueInfo['Revenue ($M)']).str.replace(',','').astype(float)

CountryRevenueInfo.to_csv(os.path.join("..","..","data","rawData","Consumer_Electronics_Sales.csv"))
```
