# This is the plot code
I restructered the data, and plotted the countries with max increase and decrease happiness score change, and two countries with min score change. I then plotted top countries and bottom contries of 2018 happiness score with all the provided factors. Plots for consumer electronics per capita for 2017 and 2018 are also included, as well as a correlation plot and a bubble plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os
path = os.path.join('..','data','cleanData','almost_complete.csv')
df=pd.read_csv(path)
# df=df.drop(['Unnamed: 0'],axis=1)
df1=df.dropna(subset=['Country_x'])
df1.rename(columns = {'Life Ladder':'Happiness Score','Country_x':'Country'}, inplace = True) 
df1['Electronic Stats per Capita']=df1['Revenue ($M)']/df['Population']*1000000
df1
```

```python
select18=df1[df1['Year']==2018]
select18=select18.set_index('Country')
select18['Happiness Score'].values
```

```python
#groupby the original dataframe to get the happiness score change
gr=df1.groupby('Country')
happiness_first=gr['Happiness Score'].first()
happiness_last=gr['Happiness Score'].last()
scores=pd.merge(pd.DataFrame(happiness_first),pd.DataFrame(happiness_last), on='Country',how='left')
scores.rename(columns = {'Happiness Score_x':'First Happiness Score Reported','Happiness Score_y':'Lastest Happiness Score Reported'}, inplace = True) 
scores['Score change during 2008-2018(%)']=(scores['Lastest Happiness Score Reported']-scores['First Happiness Score Reported'])/scores['Lastest Happiness Score Reported']*100
scores
```

```python
#find the max score change countries and make plots
scores.nlargest(3,['Score change during 2008-2018(%)'])
```

```python
Cam=df1[df1['Country']=='Cambodia']
Cambodia=Cam[['Year','Happiness Score','Log GDP per capita','Poverty Value (1-4)','Electronic Stats per Capita','Social support','Healthy life expectancy at birth','Freedom to make life choices','Generosity','Perceptions of corruption','Positive affect','Negative affect','Confidence in national government']]
fig, (ax1, ax2, ax3,ax4) = plt.subplots(4, sharex=True,figsize=(12,8))
fig.suptitle("Cambodia Score change Over Time", fontsize=16, fontweight="bold")

ax1.plot(Cambodia['Year'], Cambodia['Freedom to make life choices'], linewidth=1, marker="o")
ax1.set_ylabel("Freedom to make life choices")
ax1.set_ylim(0,1)

ax2.plot(Cambodia['Year'], Cambodia['Log GDP per capita'], linewidth=1, marker="o", color="r")
ax2.set_ylabel("Log GDP per capita")
# ax2.set_ylim(-1,0.5)

ax3.plot(Cambodia['Year'], Cambodia['Confidence in national government'], linewidth=1, marker="o", color="g")
ax3.set_ylabel("Social support")
ax3.set_ylim(0,1)

ax4.plot(Cambodia['Year'], Cambodia['Happiness Score'], linewidth=1, marker="o", color="b")
ax4.set_ylabel("Happiness Score")
plt.savefig("Images/Cambodia Score change.png")
```

```python
import seaborn as sns
import numpy as np
correlational_data = df18[['Happiness Score','Log GDP per capita','Poverty Value (1-4)','Social support','Healthy life expectancy at birth',
                             'Freedom to make life choices','Electronic Stats per Capita','Generosity','Perceptions of corruption',]]  
f,ax = plt.subplots(figsize=(10, 8))
corr = correlational_data.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True)
            ,square=True, ax=ax)
plt.savefig("Images/Corr relations.png")
```

```python
import chart_studio.plotly as py
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot
import plotly
init_notebook_mode(connected=True)
size  = df18['Happiness Score']*6
color = df18['Social support']
data = [
    {
        'y': df18['Freedom to make life choices'],
        'x': df18['Log GDP per capita'],
        'mode': 'markers',
        'marker': {
            'color': color,
            'size': size,
            'showscale': True
        },
        "text" :  df18.Country    
    }
]
layout = dict(title='Freedom vs Happiness Rank in 2018:',
              xaxis=dict(title='Log GDP per capita'),
              yaxis=dict(title='Freedom'))
fig = go.Figure(data=data,layout=layout)
fig.update_layout(
        title={
            'text': 'Freedom vs Happiness Rank in 2018: <br> Size=Happiness Score, Color=Social support',
            'x': 0.5,
            'xanchor': 'center'
        })
iplot(fig)
plotly.offline.plot(fig, filename='fig.html')
```
